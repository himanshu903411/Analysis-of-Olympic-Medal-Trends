```python
import pandas as pd

# Load dataset
df = pd.read_csv("Summer_Olympic_medallists_1896-2008.xlsx - ALL MEDALISTS.csv")
df.head()

# Module 1: Country-Wise Medal Distribution

top_countries = df['NOC'].value_counts().head(10)
top_countries.plot(kind='bar', color='gold')

# Module 2: Athlete-Level Performance Density

top_athletes = df['Athlete'].value_counts().head(10)
top_athletes.plot(kind='barh', color='silver')



# Module 3: Temporal Gender Analysis

gender_trend = df.groupby(['Edition', 'Gender'])['Athlete'].count().reset_index()
import seaborn as sns
sns.lineplot(data=gender_trend, x='Edition', y='Athlete', hue='Gender')

# Module 4: Sport-Specific Medal Trends
sport_medals = df.groupby(['Sport', 'Medal']).size().unstack().fillna(0)
sport_medals.plot(kind='bar', stacked=True, figsize=(12,6))



```
